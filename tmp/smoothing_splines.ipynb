```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline

# Generate sample data
x = np.linspace(0, 10, 100)
y = np.sin(x) + np.random.normal(scale=0.2, size=x.shape)

# Fit a smoothing spline
spline = UnivariateSpline(x, y, s=3)  # s is the smoothing factor
y_smooth = spline(x)

# Plot the original data and the smoothed curve
plt.figure(figsize=(10, 6))
plt.scatter(x, y, label='Noisy Data', color='gray', alpha=0.6)
plt.plot(x, y_smooth, label='Smoothing Spline', color='red', linewidth=2)
plt.title('Smoothing Splines Example')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
